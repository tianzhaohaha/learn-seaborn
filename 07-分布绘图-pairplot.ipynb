```python
import seaborn as sns
import pandas as pd
```

```python
iris = pd.read_csv("dataset/iris.csv")
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,vars=["sepal_length","sepal_width"])#选取想要的信息绘制图
```

```python
#添加kde曲线和回归曲线
sns.pairplot(iris,vars=["sepal_length","sepal_width"],kind="reg",diag_kind="kde")
```
