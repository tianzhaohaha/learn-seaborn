```python
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri)#外面一层表示的是置信区间，也可以删除ci=None
```

```python
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri,ci=None)
```

```python
#想对横轴不同类别进行区分，用hue
sns.relplot(x="timepoint",y="signal",kind="line",hue="region",data=fmri)#可以使用hue_order=["frontal","parietal"]对线的颜色顺序进行自定义
```

```python
#使用col进行列分割
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri,hue="region",col="event")#col="event"的含义是根据event再对图进行区分绘制

```

```python
sns.relplot(x="timepoint",y="signal",kind="line",data=fmri,hue="region",col="event",style="region")#style对绘制的线进行区分
```

```python
#注意：relplot进行绘制的时候，应该对x，y指定csv文件具体的列，而不是具体的值
```
