```python
import seaborn as sns
import pandas as pd
```

```python
tips = pd .read_csv("dataset/tips.csv")
tips.head()
```

```python
#实用relplot绘制tip和total-bill的关系
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
#为relplot添加hue参数,hue参数作用是为某一项参数添加颜色区分
sns.relplot(x="total_bill",y="tip",hue="day",data=tips)
```

```python
sns.relplot(x="total_bill",y='tip',hue="day",col="time",data=tips)#col的意思是根据col数据类别数绘制图，一个类别绘制一列
```

```python
sns.relplot(x="total_bill",y='tip',hue="day",col="time",row="sex",data=tips)
```

```python
#指定具体多少列，不想把星期几绘制到一个图上,col_wrap多少列自动换行
sns.relplot(x="total_bill",y="tip",col="day",col_wrap=2,data=tips)
```
