```python
import seaborn as sns
import pandas as pd
```

```python
#有两个函数来做线性回归绘图，一个是regplot一个是lmplot
tips = pd.read_csv("dataset/tips.csv")
```

```python
sns.regplot(x=tips['total_bill'],y=tips["tip"],data=tips)#regplot的x，y既可以是datadrame的一个列，也可以是numpy的一个数组
#regplot的data可以不传，但这样的话x，y必须指定数据
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips)
```
